```python
import pandas as pd
import numpy as np
```

# handling sentiment analysis

```python
df = pd.read_csv('./lg_sa_bert.csv')
```

```python
df
```

```python
# convert time column to datetime format
df['Time'] = pd.to_datetime(df['Time'], format='%Y/%m/%d %H:%M')

# create a new dataframe with dates as index
dates_df = pd.DataFrame(index=pd.date_range(start=df['Time'].min().date(), end=df['Time'].max().date(), freq='D'))

# initialize count columns to 0
dates_df['positive'] = 0
dates_df['negative'] = 0
dates_df['neutral'] = 0
dates_df['none'] = 0
dates_df['count'] = 0

# loop through the original dataframe and increment counts for each date
for index, row in df.iterrows():
    date = str(row['Time'].date())
    sentiment = row['Sentiment']
    if sentiment == 'positive':
        dates_df.loc[date, 'positive'] += 1
    elif sentiment == 'negative':
        dates_df.loc[date, 'negative'] += 1
    elif sentiment == 'neutral':
        dates_df.loc[date, 'neutral'] += 1
    else:
        dates_df.loc[date, 'none'] += 1
    dates_df.loc[date, 'count'] += 1
```

```python
dates_df
```

```python
dates_df.to_csv('./processed_day_level_sa_bert.csv')
```

# handling aspect-based sentiment analysis

```python
df = pd.read_csv('./lg_absa.csv', encoding='ISO-8859-1')
print(df.columns)
df
```

```python
# Drop columns starting with 'Unnamed'
df = df.loc[:, ~df.columns.str.startswith('Unnamed')]

# Convert the 'time' column to datetime format
df['time'] = pd.to_datetime(df['time'], format='%Y/%m/%d %H:%M')


df
```

```python
df.loc[0]
```

```python
# Create a new dataframe with dates as index
dates_df = pd.DataFrame(index=pd.date_range(start=df['time'].min().date(), end=df['time'].max().date(), freq='D'))

# Initialize count columns to 0
sentiments = ['positive', 'negative', 'neutral', 'none']
for col in df.columns[3:]:
    for sentiment in sentiments:
        dates_df[f'{col}_{sentiment}'] = 0

dates_df['count'] = 0
        
# Loop through the original dataframe and increment counts for each date
for index, row in df.iterrows():
    date = str(row['time'].date())
    for col in df.columns[3:]:
        sentiment = row[col]
#         print(row,col,row[col])
        if sentiment == 'positive':
            dates_df.loc[date, f'{col}_positive'] += 1
        elif sentiment == 'negative':
            dates_df.loc[date, f'{col}_negative'] += 1
        elif sentiment == 'neutral':
            dates_df.loc[date, f'{col}_neutral'] += 1
        else:
            dates_df.loc[date, f'{col}_none'] += 1
    dates_df.loc[date, 'count'] += 1
```

```python
dates_df
```

```python
dates_df.to_csv('./processed_day_level_absa.csv')
```
